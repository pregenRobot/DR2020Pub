The following notebook is going to carry out feature selection with the chi squared test

Null Hypothesis: Two variables are independent

Alternate Hypothesis: Two variables are dependent

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data_filepath = "../data/UmeaSiblingBundlingMLDistances-Top100.csv"

all_data = pd.read_csv(data_filepath)

## Dataframe of features
data = all_data.iloc[:,0:-2]

data.head()
```

```python
## Use scikit learn's implementation

from sklearn.feature_selection import chi2
from sklearn.feature_selection import SelectKBest
```

Carrying out chi squared test (and retrieving the p value) between each column and every other column. (And trying to "predict" each column based on every other column)

```python
## Can also use chi values
row_p_values = []

for i in range(len(data)):
    ## chi2([Col0,Col1 ... Col(i-1),Col(i+1),Col(i+2)...],[Col(i)])
    chi_scores = chi2(pd.concat([data.iloc[:,:i],data.iloc[:,i+1:]],axis = 1),data.iloc[:,i])
    row_p_values.append(chi_scores)
    print(i)
```

```python
chi2_selector = SelectKBest(chi2,k=2)
data_kbest = chi2selector.fit_transform(data,all_data.iloc[:,-2:-1])
```
