First of all, the positive correlation we observed between columsn 1,2,3, and 4 in Notebook 1 looks extremely weak when we are using the whole dataset.

```python
f = plt.figure(figsize=(20,20))

plt.matshow(data.corr(), fignum=f.number)
plt.colorbar()
```

Unlike the pearson correlation matrix heatmapin Notebook 1, the octate pattern is a lot more profound in this one

```python
octate = []

column_names = "00,01,02,03,04,05,06,07,08,09,10,11,12,13,14".split(",")
column_index = 0

for i in range(len(data.columns)):
    if(int(i/8)==0):
        octate.append(pd.DataFrame())
    
    dataframe = octate[i%8]
    dataframe = dataframe.assign(e=pd.Series(data.iloc[:,i]).values)
    dataframe.rename(columns={"e":column_names[int(i/8)]},inplace=True)
    octate[i%8] = dataframe
```

```python
f = plt.figure()

f, axes = plt.subplots(nrows=4,ncols=2)

f.set_figwidth(20)
f.set_figheight(20)
f.tight_layout()

counter = 0
for i in range(4):
    for ii in range(2):
        sc = axes[i][ii].imshow(octate[counter].corr())
        counter+=1
```

Looks like the octate correlation still seems to be strong between the columns. Let's try drawing a scatter plot of the columns in the first octate group.

```python
f = plt.figure()
numOfColumns = 15
f, axes = plt.subplots(nrows = numOfColumns, ncols = numOfColumns)
f.set_figwidth(20)
f.set_figheight(20)
f.tight_layout()

for i in range(numOfColumns):
    for ii in range(numOfColumns):
        sc = axes[i][ii].scatter(octate[0].iloc[:,i],octate[0].iloc[:,ii])
        axes[i][ii].set_xlabel(octate[0].columns[i])
        axes[i][ii].set_ylabel(octate[0].columns[ii])
```

Let's test for the second assumption:

2. Checking for Homoscedasticity using Levene tests (FAIL)

```python
from scipy.stats import levene

levenedata = []

for i in range(len(data.iloc[0])):
    singlerowlevene = []
    print(str(i))
    for ii in range(len(data.iloc[0])):
        singlerowlevene.append(levene(data.iloc[:,i],data.iloc[:,ii]).pvalue)
        
    levenedata.append(singlerowlevene)
    
plt.imshow(levenedata)
```

Let's try removing the diagonall line to focus on the values we care about

```python
levenedatanp = np.array(levenedata)
np.fill_diagonal(levenedatanp,0)

plt.imshow(levenedatanp*10000000000000000)
```

The above heatmap displays the p-value of the levene test carried out between each column in the dataset.

The lower the p-value is, the more the dataset violates the Levene Test's Null Hypothesis: the groups being compared all have equal population variances

Let's see the heatmap of the correlation matrix using spearman and kendall rank methods as well.  These rank correlation methods assume ordinal data (categorical data where each category is representeed by a number and the number is in order of significance). Because we noticed some straight lines in the scatter plot(above), there is a possibility that the values in the columns are categorical (and discrete)

**Spearman**

```python
f = plt.figure()

f, axes = plt.subplots(nrows=4,ncols=2)

f.set_figwidth(20)
f.set_figheight(20)
f.tight_layout()

counter = 0
for i in range(4):
    for ii in range(2):
        sc = axes[i][ii].imshow(octate[counter].corr(method="spearman"))
        counter+=1
```

**Kendall**

```python
f = plt.figure()

f, axes = plt.subplots(nrows=4,ncols=2)

f.set_figwidth(20)
f.set_figheight(20)
f.tight_layout()

counter = 0
for i in range(4):
    for ii in range(2):
        sc = axes[i][ii].imshow(octate[counter].corr(method="kendall"))
        counter+=1
```
