I will try calculating the pearson correlation value between every column in its respective octate group, in the next cell.

Another thing that we can observe is the fact that column indexes from 108 to 111 and 117 to 119 are all blank. After opening the csv table to see the full data, I have noticed that all the values are 0, suggesting that we can either completely remove these columns or the fact that the table was ascendingly sorted according to these columns and the non-zero values are towards the end of the file ( the csv file used in this notebook has sampled 100 values from the whole dataset) Further analysis is required.

```python

octate = []

column_names = "00,01,02,03,04,05,06,07,08,09,10,11,12,13,14".split(",")
column_index = 0

for i in range(len(data.columns)):
    if(int(i/8)==0):
        octate.append(pd.DataFrame())
    
    dataframe = octate[i%8]
    dataframe = dataframe.assign(e=pd.Series(data.iloc[:,i]).values)
    dataframe.rename(columns={"e":column_names[int(i/8)]},inplace=True)
    octate[i%8] = dataframe
    ## Following print statements is for debugging
    print("=======")
    print("Length: " + str(len(octate)))
    print(octate)
```

```python
f = plt.figure()

f, axes = plt.subplots(nrows=4,ncols=2)

f.set_figwidth(20)
f.set_figheight(20)
f.tight_layout()

counter = 0
for i in range(4):
    for ii in range(2):
        sc = axes[i][ii].imshow(octate[counter].corr())
        counter+=1
```

From top to bottom, left to right, the above heatmaps represent the pearson correlation matrix between each column and every 8 next columns. Compared to the first large heatmap, these heatmaps have many more correlating columns.

Vocab Research:

Linearity -- Straight line relationship between each of the two variables

Homoscedasticity -- data is equally distributed about the regression line (having the same population variance)

Regression Line -- line of best fit

Non-parametric test -- based on either being distribution-free or having a specific distribution but with the distribution's parameter unspecified

Concordant -- ordered in the same way

Discordant -- ordered differently

Ordinal Data -- Order of the values is what's important and significant
Ex: 1 - Very unhappy 2 - Unhappy 3 - OK 4 - Happy 5 - Very Happy

Monotonically Related -- as x gets bigger, so does f(x) and as x gets smaller, so does f(x)
